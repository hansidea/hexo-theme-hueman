## 範例用: 抓取Google自動完成關鍵字 
Created by [數位引擎](https://www.darrenhuang.com/google-autocomplete-keyword.html)


```
keyword='\u97D3\u570B\u745C' #@param {type:"string"}
```

```
!pip3 install fake_useragent
import requests
import json
from fake_useragent import UserAgent
import pandas as pd
import time
```

```
def gSuggest(query):
  query.replace(" ", "+")
  hl="zh-TW"
  gl="tw"
  url = "http://suggestqueries.google.com/complete/search?output=firefox&q={}&hl={}&gl={}".format(query,hl,gl)
  ua = UserAgent(verify_ssl=False) 
  headers = {"user-agent": ua.chrome}

  response = requests.get(url, headers=headers, verify=True)
  results = json.loads(response.text)
  return results

def word_group(results):
  data = []
  for word in results[1]:
    k = gSuggest(word)[0]
    deep_words = gSuggest(word)[1]

    for deep_word in deep_words:
      group = [results[0], k, deep_word]
      data.append(group)
  
  return data

def get_autocomp_kws(query):
  results = gSuggest(query)
  data = word_group(results)
  return data
```

```
data =[]

google_suggest = get_autocomp_kws(keyword)
data = data+google_suggest
print('處理[{}]中...\n'.format(keyword))
time.sleep(2)

print("完成\n")
df = pd.DataFrame(data, columns=['原始關鍵字', '第一層預測字', '第二層預測字'])
df
```

```
df.to_csv('output.csv', index=False, header=True, encoding='utf-8-sig')
```
